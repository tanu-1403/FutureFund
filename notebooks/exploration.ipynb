2. Import Your Modules

```python
import sys
import os
sys.path.append(os.path.abspath(".."))  # adjust path if notebook is inside a subfolder

# Import modules
from utils.goal_manager import initialize_goals, add_goal, get_goals, display_sticky_goals
from utils.helpers import calculate_monthly_sip, format_currency, generate_insight
from src.financial_engine import future_goal_value, required_sip
from src.scenario_simulator import simulate_delay
from src.projections import investment_projection
from src.monte_carlo import monte_carlo_simulation
from src.life_simulator import simulate_life
from src.ai_advisor import generate_advice
```

3. Initialize Goals

```python
initialize_goals()

# Add example goals
add_goal(name="Buy Car", cost=800000, years=5, goal_type="Car", note="Need a sedan")
add_goal(name="House", cost=5000000, years=10, goal_type="House", note="Down payment")
add_goal(name="Higher Education", cost=2000000, years=7, goal_type="Education", note="Masters abroad")

display_sticky_goals()
```

4. Calculate Monthly SIPs for Goals

```python
for g in get_goals():
    sip = calculate_monthly_sip(g['future_cost'], g['years'])
    print(f"Goal: {g['name']} | Future Cost: {format_currency(g['future_cost'])} | Monthly SIP: {format_currency(sip)}")
```

5. Simulate Delay Impact

```python
delay_results = simulate_delay("Buy Car", 800000, 5, delay=2)
print(delay_results['insight'])
```

6. Project Investment Growth

```python
goal_sip = calculate_monthly_sip(5000000, 10)
portfolio = investment_projection(goal_sip, annual_return=0.08, years=10)

import matplotlib.pyplot as plt

months = [p['month'] for p in portfolio]
values = [p['nominal'] for p in portfolio]

plt.figure(figsize=(10,5))
plt.plot(months, values, label="Nominal Portfolio")
plt.title("Investment Projection for House")
plt.xlabel("Month")
plt.ylabel("Portfolio Value")
plt.legend()
plt.show()
```

7. Run Monte Carlo Simulation

```python
mc = monte_carlo_simulation(sip=goal_sip, years=10, goal_value=5000000, simulations=1000)
print(f"Mean Portfolio: {format_currency(mc['mean'])}")
print(f"Probability of reaching goal: {mc['success_prob']:.1f}%")
```

8. Lifetime Wealth Simulation

```python
wealth_history = simulate_life(income=100000, savings=0.3, growth=0.08, years=30, monthly=True)

plt.figure(figsize=(10,5))
plt.plot([h['date'] for h in wealth_history], [h['real_value'] for h in wealth_history])
plt.title("Lifetime Wealth Projection (Inflation-adjusted)")
plt.xlabel("Date")
plt.ylabel("Real Wealth")
plt.show()
```
